# Notes: [Integrate-And-Fire Model](https://neuronaldynamics.epfl.ch/online/Ch1.S3.html)

We describe the critical voltage for spike initiation by a formal **threshold** $\theta$.If the **voltage** $u_i(t)$ (that contains the summed effect of all inputs) reaches $\theta$ from below, we say that **neuron** $i$ fires a **spike**. The moment of threshold crossing defines the **firing time** $t_i^{(f)}$

the simplest model in the class of integrate-and-fire models using the following two ingredients: 
* a **linear differential equation** to describe the evolution of the **membrane potential**
* a **threshold** for **spike firing**

This model is called the **Leaky Integrate-and-Fire Model**. 

The variable $u_i$ describes the **momentary value** of the **membrane potential** of **neuron** $i$.
In the absence of any input, the potential is at its **resting value** $u_{rest}$

## Integration of Inputs

**momentary voltage** $u_i(t)−u_{rest}$

**input current** $I(t)$

**cell membrane** acts like a capacitor of **capacity** $C$

- additional electrical charge $q=\int I(t^{'})\mathrm{d}t^{'}$has to go somewhere

**cell membrane** can be characterized by a finite leak __resistance__ $R$.

- the charge will, over time, slowly leak through the cell membrane. 

**law of conservation**
$$I(t)=I_R+I_C$$

for $I_R=u_R/R$ where $u_R=u-u_{rest}$

for $I_C=\mathrm{d}q/\mathrm{d}t=C \mathrm{d}u/\mathrm{d}t$

we have
   $$I(t)=\frac{u(t)-u_{rest}}{R}+C \frac{\mathrm{d}u}{\mathrm{d}t}$$
   
with **time constant** $\tau_m=R C$
    $$\tau_m\frac{\mathrm{d}u}{\mathrm{d}t}=-\left[u(t)-u_{rest}\right]+R I(t)$$
    
This is the **leaky integration** equation
    
**membrane potential** of the neuron $u$

**membrane time constant** of the neuron $\tau_m$

Two assumption:
* $u(t_0)=u_{rest}+\Delta u$
* $I(t)=0$, for $t>t_0$

$$u(t)-u_{rest}=\Delta u\ \exp\left(-\frac{t-t_0}{\tau_m} \right)\text{ for } t>t_0 $$

## Pulse Input

Suppose that the **passive membrane** is stimulated by($t_0=0$):
* $I(t)=I_0$, $t\in\left[0,\Delta\right]$ 
* $u(0)=u_{rest}$

we have $0<t<\Delta$
$$u(t)=u_{rest}+R I_0\left[1- exp(-\frac{t}{\tau_m})\right]$$

$t\rightarrow\infty$, $u(\infty)=u_{rest}+R I_0$

the **total membrane voltage** is $u_{rest}+R I_0$.

## The Threshold for Spike Firing

**firing time** refers to the moment when a given neuron emits an action potential $t^{(f)}$

The firing time $t^{(f)}$ in the leaky integrate-and-fire model is defined by a threshold criterion:

$$t^{(f)}:u(t^{f})=\theta$$


the **firing time** is noted and immediately after $t_{(f)}$ the potential is reset to a new value $u_r<\theta$,
$$\text{lim}_{\delta->0;\delta>0} u(t^{(f)}+\delta)=u_r$$

This is the **reset** function

The combination of **leaky integration** and **reset** defines **the leaky integrate-and-fire model**

the firing times of neuron $i$ we write $t_i^{(f)}$ where $f=1,2,...$ is the label of the spike.We denote the spike train of a neuron i as the sequence of firing times

$$S_i(t)=\sum_f\delta(t-t_i^{(f)})$$

## Time-dependent Input(*)

In the absence of a threshold, the linear differential equation (**leaky integration equation**)has a solution

$$u(t)=u_{rest}+\frac{R}{\tau_m}\int_0^\infty\exp{-\frac{s}{\tau_m}}I(t-s)\mathrm{d}s$$

$I(t)$ can be
$$I(t)=I_0\ sin(\omega t)$$
or
$$I(t)=I_0\ \Theta(t)\text{  , where }\Theta(t)=\begin{cases} 
      0 & t\leq0 \\
      1 & t>0
   \end{cases}$$

After adding a threshold $\theta$ into the Equation

$$I_r=-q_r\sum_f\delta(t-t^{(f)})=-C(\theta-u_r)S(t)$$

Each time the membrane potential hits the threshold, the variable $u$ is **reset** from $\theta$ to $u_r$. In the electrical circuit diagram, the reset of the potential corresponds to removing a **charge** $q_r=C(\theta-u_r)$ from the capacitor. Therefore, the **reset** corresponds to a short current pulse $I_r=-q_r\sum_f\delta(t-t^{(f)})$ at the moment of the firing t(f).

The short current pulse corresponding to the **'discharging'** is treated mathematically just like any other time-dependent input current.
The **total current** $I(t)+I_r(t)$ 

$$u(t)=u_{rest}+\sum_f(u_r-\theta)\exp\left(-\frac{t-t^{(f)}}{\tau_m}\right)\\+\frac{R}{\tau_m}\int_0^\infty\exp\left(-\frac{s}{\tau_m}\right)I(t-s)\mathrm{d}s$$

where $t^{(f)}=\left{t|u(t)=\theta \right}$

a short input pulse at time $t^{'}$ causes at time $t$ a response of the membrane proportional to $\exp\left({-\frac{t-\tau^{'}}{\tau_m}}\right)$, sometimes called the **impulse response function** or **Green’s function**

## Linear Differential Equation vs. Linear Filter: Two Equivalent Pictures (*)

rewrite the total current eqaution shown above:

$$u(t)=\int_0^\infty\eta(s)S(t-s)\mathrm{d}s+
\int_0^\infty \kappa(s) I(t-s)\mathrm{d}s$$

$\eta(s)=\frac{1}{C}\exp\left(-\frac{s}{\tau_m}\right)$ and $\kappa(s)=\frac{1}{C}\exp\left(-\frac{s}{\tau_m}\right)$

## Periodic drive and Fourier transform (*)

$$\hat{f}(\omega)=\int_{-\infty}^\infty f(t) e^{-i\omega t}\mathrm{d}t=|\hat{f}(\omega)|e^{i\phi_f(\omega)}$$

$|\hat{f}(\omega)|$ and $\phi_f(\omega)$ are called **amplitude** and **phase** of the Fourier transform

The phase $\phi_\kappa(\omega)$ of the Fourier-transformed linear filter $\kappa$ corresponds to phase shift between input and output or, to say it differently, a delay
$\Delta=\phi_\kappa/\omega=\phi_\kappa T/2\pi$

### Question: minimal current (simulation)

```python
V_REST = -70*b2.mV
V_RESET = -65*b2.mV
FIRING_THRESHOLD = -50*b2.mV
MEMBRANE_RESISTANCE = 10. * b2.Mohm
MEMBRANE_TIME_SCALE = 8. * b2.ms
ABSOLUTE_REFRACTORY_PERIOD = 2.0 * b2.ms
```

```python
LIF.MEMBRANE_TIME_SCALE?
```

For the default neuron parameters (see above) compute the minimal amplitude ***i_min*** of a step current to elicitate a spike.

```python
from neurodynex.leaky_integrate_and_fire import LIF
print("rest potential: {}".format(LIF.V_REST))
print("reset potential: {}".format(LIF.V_RESET))
print("firing threshold: {}".format(LIF.FIRING_THRESHOLD))
print("membrane resistance: {}".format(LIF.MEMBRANE_RESISTANCE))
print("time constant: {}".format(LIF.MEMBRANE_TIME_SCALE))
print("absolute refractory period: {}".format(LIF.ABSOLUTE_REFRACTORY_PERIOD))
```

```python
# t0=5,t=100
# 95=100-5
i_min=(FIRING_THRESHOLD-V_REST)/(MEMBRANE_RESISTANCE*(1-b2.exp(-95*b2.ms/(MEMBRANE_TIME_SCALE))))
i_min
```

### Question: minimal current (simulation)

```python
import brian2 as b2
from neurodynex.leaky_integrate_and_fire import LIF
from neurodynex.tools import input_factory


# create a step current with amplitude= i_min
step_current = input_factory.get_step_current(t_start=5, t_end=100, unit_time=b2.ms,
                                              amplitude= i_min)  # set i_min to your value

# run the LIF model.
# Note: As we do not specify any model parameters, the simulation runs with the default values
(state_monitor,spike_monitor) = LIF.simulate_LIF_neuron(input_current=step_current, simulation_time = 100 * b2.ms)

# plot I and vm
plot_tools.plot_voltage_and_current_traces(state_monitor, step_current, title="min input",
                                           firing_threshold=LIF.FIRING_THRESHOLD)
print("nr of spikes: {}".format(spike_monitor.count[0]))  # should be 0
```

```python
# object 3
# time consuming
list_current=[float(i)*b2.namp for i in range(0,101,1)]
freq_spikes=[]
for a in list_current:
    step_current = input_factory.get_step_current(t_start=5, t_end=500, unit_time=b2.ms,amplitude=a)  # set i_min to your value
    (state_monitor,spike_monitor)=LIF.simulate_LIF_neuron(
        input_current=step_current,
        simulation_time = 500 * b2.ms,
        abs_refractory_period=3*b2.ms)
    freq_spikes.append(1/(spike_monitor.count[0]/0.5))
```

```python
plt.scatter(x=np.arange(101),y=freq_spikes)
plt.xlabel('i_min(nA)')
plt.ylabel('frequency(Hz)')
```
