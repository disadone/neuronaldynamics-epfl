Using **Kirchhoff’s laws** we find equations that relate the **voltage** $u(x)$ across the membrane at **location** $x$ with longitudinal and transversal currents. 
> **Kirchhoff's current law (1st Law)** states that current flowing into a node (or a junction) must be equal to current flowing out of it. This is a consequence of charge conservation. 
>
>**Kirchhoff's voltage law (2nd Law)** states that the sum of all voltages around any closed loop in a circuit must equal zero. 

First, a longitudinal current $i(x)$ passing through the dendrite causes a voltage drop across the **longitudinal resistor** $R_L$ according to **Ohm’s law**,

$$u(t,x+\mathrm{d}x)-u(t,x)=R_L i(t,x)$$

the transversal current that passes through the RC-circuit is given by $C \partial u(t,x)/\partial t+\sum_{ion}I_{ion}$ where ***the sum runs over all ion channel types present in the dendrite***. **Kirchhoff’s law** regarding the conservation of current at each node leads to

$$i(t,x+\mathrm{d}x)-i(t,x)=C \frac{\partial}{\partial t} u(t,x)+\sum_{ion}I_{ion}-I_{ext}(t,x)$$

The values of the **longitudinal resistance** $R_L$, the **capacity** $C$, the **ionic currents** as well as the **externally applied current** can be expressed in terms of specific quantities per unit length $r_L$, $c$, $i_{ion}$ and $i_{ext}$, respectively

$R_L=r_L\mathrm{d}x$,
$C=c\mathrm{d}x$,
$I_{ext}(t,x)=i_{ext}(t,x)\mathrm{d}x$,
$I_{ion}(t,x)=i_{ion}(t,x)\mathrm{d}x$

Substituting these expressions in above formulas, dividing by $\mathrm{d}x$, and taking the limit $\mathrm{d}x\rightarrow 0$ leads to

$$\frac{\partial}{\partial t} u(t,x)=r_L i(t,x)$$
$$\frac{\partial}{\partial t} i(t,x)=
c \frac{\partial}{\partial t} u(t,x)+\sum_{ion}i_{ion}(t,x)-i_{ext}(t,x)
$$

Derivative of fisrt equation w.r.t $x$

$$\frac{\partial^2}{\partial x^2}u(t,x)=c\ r_L \frac{\partial}{\partial t} u(t,x)+r_L \sum_{ion}i_{ion}(t,x)-r_L i_{ext}(t,x)$$

This is called **the general cable equation**.

## Derivation of the Cable Equation

### Example: Cable equation for passive dendrite



For simplicity, the dendrite is sometimes considered as passive. This means that the current density follows Ohms law

$$\sum_{ion} i_{ion}=g_l(u-E_l)$$

where $g_l=1/r_T$

$g_l=1/r_T$ is the **leak conductance per unit length**

$E_l$ is the **leak reversal potential**


**characteristic length scale** $\lambda^2=r_T/r_L$

**memebrane time constant** $\tau=r_T c$

$$
\lambda^2 \frac{\partial^2}{\partial x^2}u(t,x)= \tau \frac{\partial}{\partial t} u(t,x)+\left[u(t,x)-E_l\right]-r_T i_{ext}(t,x)$$

After a transformation to unit-free coordinates and a rescaling of the current and voltage variables

$$x\rightarrow\hat{x}=x/\lambda,t\rightarrow\hat{t}=t/\tau$$

$$i\rightarrow\hat{i}=\sqrt{r_T r_L}i\ ,\ i_{ext}\rightarrow\hat{i}_{ext}=r_T i_ext\ ,\ u\rightarrow\hat{u}=u-E_l$$

we get

$$\frac{\partial}{\partial t}u(t,x)= \frac{\partial^2}{\partial x^2} u(t,x)-u(t,x)+i_{ext}$$

**cable equation in an elegant unit-free form**

### Example: Stationary solutions of the cable equation

for solutions with $\partial u(t,x)/\partial t=0$

$$\frac{\partial^2}{\partial x^2}u(t,x)-u(t,x)=-i_{ext}(t,x)$$

the general solution to the homogenous equation with $i_{ext}(t,x)\equiv0$ is

$$u(t,x)=c_1 \mathrm{sinh}(x)+c_2 \mathrm{cosh}(x)$$

> $sinh x=\frac{e^x-e^{-x}}{2}$
>
> $cosh x=\frac{e^x+e^{-x}}{2}$

for a stationary input current $i_{ext}(t,x)=\delta(x)$ localized at $x=0$ and boundary conditions $u(\pm\infty)=0$, we find

$$u(t,x)=\frac{1}{2}e^{-|x|}$$

note that
$r_L=R_L/(\pi\rho^2)$ and $r_T=R_T/(2\pi\rho)$

## Green’s Function of the passive Cable

Suppose a short current pulse $i_{ext}(t,x)$ is injected at time $t=0$ at location $x=0$. As we will show below, the time course of the voltage at an arbitrary position $x$ is given by

$$u(t,x)=\frac{\Theta(t)}{\sqrt{4\pi t}}\exp\left[-t-\frac{x^2}{4 t}\right]\equiv G_{\infty}(t,x)$$

with $\Theta(t)$ denoting the **Heaviside function**.

> **Heaviside function**
> $$H(x)=\begin{cases}
0 & x<0\\
\frac{1}{2}& x=0\\
1& x>0
\end{cases}
$$
>$$\frac{\mathrm{d}}{\mathrm{d}x}H(x)=\delta(x)$$

### Derivation of the Green's function

$i_{ext}=\delta(t)\delta(x)$ for a impulse at $x=0$ and $t=0$

$$\frac{\partial}{\partial t}u(t,x)-\frac{\partial^2}{\partial x}u(t,x)+u(t,x)=\delta(t)\delta(x)$$

(A little special)Fourier transformation w.r.t. $x$

$$u(t,k)=\mathcal{F}\left[u(t,x)\right]=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty u(t,x)e^{-i x k}\mathrm{d}x$$


$$\mathcal{F}\left[\frac{\partial}{\partial t}u(t,x) \right]
-\mathcal{F}\left[\frac{\partial^2}{\partial x}u(t,x)\right]
+\mathcal{F}\left[u(t,x)\right]=\delta(t)\mathcal{F}\left[\delta(x)\right]$$

for $\frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty \delta(x-a) e^{-i x k}\mathrm{d}x= \frac{1}{\sqrt{2\pi}} e^{-i k a}$,$a=0$,hence it equals $\frac{1}{\sqrt{2\pi}}$

$$\frac{\partial}{\partial t}\mathcal{F}\left[u(t,x) \right]
-i^2 k^2\mathcal{F}\left[u(t,x)\right]
+u(t,k)=\delta(t)/\sqrt{2\pi}$$


$$\frac{\partial}{\partial t}u(t,k)+k^2 u(t,k)+u(t,k)=\delta(t)/\sqrt{2\pi}$$

By Observation,

$$u(t,k)=\exp\left[-(1+k^2) t\right]/\sqrt{2\pi}\Theta(t)$$

Apply (a little special) inverse Fourier Transformation,

$$u(t,x)=\mathcal{F}^{-1}\left[u(t,k)\right]=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^\infty u(t,k)e^{i x k}\mathrm{d}k$$

### Example : Finite cable


We consider a finite cable extending from $x=0$ to $x=L$ with sealed ends, i.e., $i(t,x=0)=i(t,x=L)=0$ or, equivalently, $\frac{\partial}{\partial x}u(t,x=0)=\frac{\partial}{\partial x} u(t,x=L)=0$.

**"mirror charges"**

letting $i(t,x=L)=0$,

$x=-x_0$ and $x=2 L+x_0$ impulse for they are symmetric in $L$

however this will damage condition $i(t,x=0)=0$

to fix this,$x=x_0$ and $x=-2 L-x_0$ is needed

however this damage condition $i(t,x=L)=0$

... so on

these impulses happens at the same time.

$$G_{0,L}(t_0,x_0;t,x)=\sum_{n=-\infty}^\infty G_{\infty}(t-t_0,x-2 n L-x_0)+G_{\infty}(t-t_0,x-2 n L+x_0)$$

in order to be **translation invariant**

$$u(t,x)=\int_{-\infty}^t\mathrm{d}t_0\int_0^L \mathrm{d}x_0 G_{0,L}(t_0,x_0;t,x)i_{ext}(t_0,x_0)$$
